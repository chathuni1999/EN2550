1. Write a program to display the squares of numbers from 1-5

```python
for i in range(1,6):
  print(i,":",i*i)
```

2. Alter the code in (1) to print the square only of non-prime numbers. Use the isprime
function in the sympy package for testing if a number is a prime.

```python
import sympy

for i in range(1,6):
  if not sympy.isprime(i):
    print(i,":",i*i)
```

3. Use a comprehension to do (1).

```python
squares = [i*i for i in range(1,6)]
for i,i2 in enumerate(squares):
  print(i+1,":",i2)
```

4. Use a comprehension to do (2).

```python
squares = [i*i for i in range(1,6) if not sympy.isprime(i)]
for i,i2 in enumerate(squares):
  print(i+1,":",i2)
```

```python
import numpy as np

A = np.array([[1,2],[3,4],[5,6]])
B = np.array([[7,8,9,1],[1,2,3,4]])

C = np.matmul(A,B)
print("Matrix Multiplication")
print(C)

D = np.array([[3,2],[5,4],[3,1]])
E = np.multiply(A,D)
print("Elment-wise Multiplication")
print(E)
```

8. Consider the following code snippet

```python
m, c = 2 , -4
N = 10
x = np.linspace(0 , N-1, N).reshape (N, 1 )
sigma = 10
y = m*x + c + np.random.normal(0 , sigma , (N, 1 ))
```

(a) Append a column of ones to x to create X.
(b) Compute the mentioned.

```python
ones = np.ones((N,1))
X = np.append(ones,x,1)
print("X")
print(X)

out1 = np.matmul(np.transpose(X),X)
out2 = np.linalg.inv(out1)
out3 = np.matmul(out2,np.transpose(X))
out4 = np.matmul(out3,y)
print("Output")
print(out4)
```

10. Use Gaussian smoothing to filter the noise in the image gal_gaussian.png shown in
Fig. 1.

```python
im = cv.imread(r'./Images/gal_gaussian.png')

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
fig,ax = plt.subplots()
ax.imshow(cv.cvtColor(im, cv.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()

# Display after Median Filtering
blur = cv.GaussianBlur(im,(5,5),0)
cv.imshow('Image',blur)
cv.waitKey(0)
cv.destroyAllWindows()

fig,ax = plt.subplots()
ax.imshow(cv.cvtColor(blur, cv.COLOR_BGR2RGB))
plt.title("Smoothened Image")
plt.show()
```
